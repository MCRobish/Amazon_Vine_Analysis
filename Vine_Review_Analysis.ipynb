```python
import pandas as pd 
```

```python
vine_df = pd.read_csv('./vine_table.csv')
vine_df.head()
```

```python
high_votes = vine_df[vine_df['helpful_votes']>20]
high_votes.head(20)
```

```python
high_votes.info()
```

```python
votes_ratio = high_votes[high_votes['helpful_votes']/high_votes['total_votes']>=0.5]
votes_ratio.head()
```

```python
votes_ratio.info()
```

```python
paid_reviews=votes_ratio[votes_ratio['vine'] == "Y"]
paid_reviews.head()
```

```python
unpaid_reviews=votes_ratio[votes_ratio['vine'] == "N"]
unpaid_reviews.head()
```

```python
#count of total reviews 
Total_paid= paid_reviews["review_id"].count()
print(f'The total number of paid reviews are: {Total_paid}')
Total_unpaid= unpaid_reviews["review_id"].count()
print(f'The total number of unpaid reviews are: {Total_unpaid}')
```

```python
#count of 5 star reviews
Paid_fivestar= paid_reviews[paid_reviews["star_rating"]==5.0].count()
paid_count= Paid_fivestar['star_rating']
print(f'The total number of paid 5-star ratings are: {paid_count}')

unpaid_fivestar= unpaid_reviews[unpaid_reviews["star_rating"]==5.0].count()
unpaid_count= unpaid_fivestar['star_rating']
print(f'The total number of unpaid 5-star ratings are: {unpaid_count}')
```

```python
#Percentage of 5 star reviews 
paid_percentage= (paid_count/Total_paid)*100
print(f'The percentage of paid reviews that are 5-stars are: {paid_percentage}%')
unpaid_percentage=(unpaid_count/Total_unpaid)*100
print(f'The percentage of unpaid reviews that are 5-stars are: {unpaid_percentage}%')
```
